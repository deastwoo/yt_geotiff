```python
import numpy as np
import yt
import yt.extensions.geotiff
from IPython.display import Image
```

```python
filename = 'example.tif'
ds = yt.load(filename)
```

```python
print(ds.field_list)
print(ds.parameters)
ds.print_stats()

slc = yt.SlicePlot(ds, 'z', '1')
# slc.save()
```

```python
image = slc.save()[0]
Image(filename=image) 
```

```python
ad = ds.all_data()
```

```python
yt.ProfilePlot(ad, "x", ('bands', '1'), weight_field=None)
```

```python
sp = ds.sphere([150000, 150000, 0], (100, 'km'))
```

```python
len(sp[('bands', '1')])
```

```python
yt.ProfilePlot(sp, "x", ('bands', '1'), weight_field="dy")
```
